# Comparing different classification model performance on the Breast Cancer dataset

The dataset contains comparable amount of data belonging to both classes avoiding any imbalance. We can therefore compare the accuracies.

```python
KNN = 0.9473684210526315
Support_Vector = 0.9532163742690059
Naive_Bayes = 0.9415204678362573
Decision_Tree = 0.9590643274853801
Random_Forrest = 0.935672514619883
XG_Boost = 0.9781021897810219
```

```python
import matplotlib.pyplot as plt

Model = ['KNN','Support_Vector','Naive_Bayes','Decision_Tree','Random_Forrest','XG_Boost']
Accuracy = [0.9473684210526315,0.9532163742690059,0.9415204678362573,0.9590643274853801,0.935672514619883,0.9781021897810219]
plt.rcParams["figure.figsize"] = (12, 8)
plt.xlim([0.85, 1])
plt.style.use('ggplot')

plt.barh(Model,Accuracy)
plt.title('Accuracy score for different classification models')
plt.ylabel('Model', fontsize = 12)
plt.xlabel('Accuracy', fontsize = 12)
plt.show()
```

### Comparing the Accuracy obtained from different models we can conclude that the XG_Boost model suits the best for the Breast Cancer Dataset
