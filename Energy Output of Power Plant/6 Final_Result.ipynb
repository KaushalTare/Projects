

# Comparing different model performances on the data set

Model Performance metric / R-squared coefficiet value for different regression models

```python
multiple_lin_reg = 0.9325315554761302
polynomial_reg = 0.9443678641003473
support_vector = 0.9480784049986258
decision_tree = 0.922905874177941
random_forrest = 0.9615908334363876
```

```python
import matplotlib.pyplot as plt
   
Model = ['multiple_lin_reg','polynomial_reg','support_vector','decision_tree','random_forrest']
r2_score = [0.9325315554761302,0.9443678641003473,0.9480784049986258,0.922905874177941,0.9615908334363876]

New_Colors = ['green','blue','purple','brown','teal']
plt.rcParams["figure.figsize"] = (12, 8)
plt.xlim([0.8, 1])
plt.barh(Model, r2_score, color = New_Colors)
plt.style.use('ggplot')
plt.title('r2_score for different regression models')
plt.xlabel('r2_score', fontsize = 14)
plt.ylabel('Model', fontsize = 14)
plt.show()
```

**Comparing the performance of different models we can conclude that the Random Forrest regressor suits the best for the Power Plant Dataset**
